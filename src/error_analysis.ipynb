```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
LABELS = {"B": "B-MUS", "PB": "B-PER", "I": "I-MUS", "PI": "I-PER", "O": "O"}     
new_labels = list(LABELS.values())
```

```python
def convert_dev_from_string_to_conll(dev_path, conll_path):
    with open(dev_path, "r") as fin:
        print(conll_path, type(conll_path))
        with open(conll_path, "w") as fout:
            for line in fin.readlines():
                try:
                    for tagged_word in line.split():
                        word, tag = tagged_word.split("-[", 1)
                        old_tag = tag.split("]", 1)[0]
                        new_tag = LABELS[old_tag]
                        fout.write(f"{word} {new_tag}\n")
                except Exception as e:
                    raise Exception(line, e)
                    
def get_set_labels(set_path):
    set_labels = []
    with open(set_path, "r") as f:
        for t in f.readlines():
            label = t.split()[-1]
            set_labels.append(label)
    return set_labels
                    
```

```python
dev_path = '../data/musicians_dataset/dev.txt'
conll_path = '../experiments/predictions/dev.conll'


convert_dev_from_string_to_conll(dev_path, conll_path)

```

```python
gold_path = '../experiments/predictions/musicians_with_hearst-1.conll'

pred_path = '../experiments/predictions/musicians_with_hearst_manual_dev-1.conll'
gold_labels = get_set_labels(gold_path)
print(gold_labels[-10:], len(gold_labels))
pred_labels = get_set_labels(pred_path)
print(pred_labels[-10:], len(pred_labels))


confusion_matrix(gold_labels, pred_labels, labels=new_labels)
```
