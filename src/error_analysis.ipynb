```python
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
LABELS = {"B": "B-MUS", "PB": "B-PER", "I": "I-MUS", "PI": "I-PER", "O": "O"}     
new_labels = list(LABELS.values())
```

```python
def convert_dev_from_string_to_conll(dev_path, conll_path):
    with open(dev_path, "r") as fin:
        with open(conll_path, "w") as fout:
            for line in fin.readlines():
                try:
                    for tagged_word in line.split():
                        word, tag = tagged_word.split("-[", 1)
                        old_tag = tag.split("]", 1)[0]
                        new_tag = LABELS[old_tag]
                        fout.write(f"{word} {new_tag}\n")
                except Exception as e:
                    raise Exception(line, e)
                    
def get_set_labels(set_path):
    set_labels = []
    with open(set_path, "r") as f:
        for t in f.readlines():
            label = t.split()[-1]
        set_labels.append(label)
    return set_labels
                    
                    
def confusion_matrix(gold_path, pred_path):
    gold = get_set_labels(gold_path)
    pred = get_set_labels(pred_path)
    return confusion_matrix(gold, pred)
```
